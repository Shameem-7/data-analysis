## Standardizing data

```python
from sklearn.preprocessing import scale
```

```python
df=df[['survived', 'pclass', 'age', 'sibsp','fare']]
```

```python
df.columns
```

```python
sc_data=scale(df)
sc_data
```

```python
scl_data=pd.DataFrame(sc_data,columns=df.columns)
scl_data
```

```python
df["fare"].mean(),df["fare"].std()
```

```python
scl_data["fare"].mean(),scl_data["fare"].std()
```

```python
df.mean(),df.std()
```

```python
scl_data.mean(),scl_data.std()
```
